# Overview
The purpose of this analysis is to visualize the synthetic beneficiary population by year.<br>
Visit Data.gov to download the data used for this analysis.

# Load Libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
from pathlib import Path
import matplotlib
import matplotlib.pyplot as plt


matplotlib.rcParams['figure.figsize'] = (12,8)
dpi=200
matplotlib.rcParams['figure.dpi']=dpi
```

# Load Data

```python
# load data
bene_2015 = pd.read_csv('data/beneficiary_2015.csv', sep="|", usecols=['BENE_ID'], low_memory=False)
bene_2016 = pd.read_csv('data/beneficiary_2016.csv', sep="|", usecols=['BENE_ID'], low_memory=False)
bene_2017 = pd.read_csv('data/beneficiary_2017.csv', sep="|", usecols=['BENE_ID'], low_memory=False)
bene_2018 = pd.read_csv('data/beneficiary_2018.csv', sep="|", usecols=['BENE_ID'], low_memory=False)
bene_2019 = pd.read_csv('data/beneficiary_2019.csv', sep="|", usecols=['BENE_ID'], low_memory=False)
bene_2020 = pd.read_csv('data/beneficiary_2020.csv', sep="|", usecols=['BENE_ID'], low_memory=False)
bene_2021 = pd.read_csv('data/beneficiary_2021.csv', sep="|", usecols=['BENE_ID'], low_memory=False)
bene_2022 = pd.read_csv('data/beneficiary_2022.csv', sep="|", usecols=['BENE_ID'], low_memory=False)
bene_2023 = pd.read_csv('data/beneficiary_2023.csv', sep="|", usecols=['BENE_ID'], low_memory=False)
bene_2024 = pd.read_csv('data/beneficiary_2024.csv', sep="|", usecols=['BENE_ID'], low_memory=False)
bene_2025 = pd.read_csv('data/beneficiary_2025.csv', sep="|", usecols=['BENE_ID'], low_memory=False)
```

```python
# add year column to each df
bene_2015.insert(1,"table_name", '2015')
bene_2016.insert(1,"table_name", '2016')
bene_2017.insert(1,"table_name", '2017')
bene_2018.insert(1,"table_name", '2018')
bene_2019.insert(1,"table_name", '2019')
bene_2020.insert(1,"table_name", '2020')
bene_2021.insert(1,"table_name", '2021')
bene_2022.insert(1,"table_name", '2022')
bene_2023.insert(1,"table_name", '2023')
bene_2024.insert(1,"table_name", '2024')
bene_2025.insert(1,"table_name", '2025')
```

```python
# union beneficiary tables
unioned_benes = pd.concat([bene_2015,bene_2016,bene_2017,bene_2018,bene_2019,bene_2020,bene_2021,bene_2022,bene_2023,bene_2024,bene_2025])
```

```python
# set style
sns.set_style("dark", {'axes.grid' : False})

# set figure size
plt.figure(figsize=(10, 5), dpi=250)

# graph that shows the synthetic bene population by year to illustrate that we have claims going out until 2025.
ax = sns.countplot(data=unioned_benes, x='table_name', color='mediumseagreen')
ax.axes.set_title('Synthetic Beneficiary Population by File Year',fontsize=18)
ax.set_xlabel("Year",fontsize=15)
ax.set_ylabel("Count of Beneficiaries",fontsize=15)
ax.xaxis.grid(False)
ax.tick_params(labelsize=13)

# label bars
for c in ax.containers:
    labels = [f'{h} ({h/unioned_benes.table_name.count()*100:0.0f}%)' if (h := v.get_height()) > 0 else '' for v in c]
    ax.bar_label(c, labels=labels, label_type='edge', fontsize=9)
```

```python
# graph limited to 2023
unioned_benes_thru_2023 = pd.concat([bene_2015,bene_2016,bene_2017,bene_2018,bene_2019,bene_2020,bene_2021,bene_2022,bene_2023])

# set style
sns.set_style("dark", {'axes.grid' : False})

# set figure size
plt.figure(figsize=(10, 5), dpi=250)

# graph that shows the synthetic bene population by year to illustrate that we have claims going out until 2025.
ax = sns.countplot(data=unioned_benes_thru_2023, x='table_name', color='mediumseagreen')
ax.axes.set_title('Synthetic Beneficiary Population by File Year',fontsize=18)
ax.set_xlabel("Year",fontsize=15)
ax.set_ylabel("Count of Beneficiaries",fontsize=15)
ax.xaxis.grid(False)
ax.tick_params(labelsize=13)

# label bars
for c in ax.containers:
    labels = [f'{h} ({h/unioned_benes_thru_2023.table_name.count()*100:0.0f}%)' if (h := v.get_height()) > 0 else '' for v in c]
    ax.bar_label(c, labels=labels, label_type='edge', fontsize=9)
```
